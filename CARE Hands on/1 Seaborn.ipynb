```python
import seaborn as sns
sns.set()
```

```python
sns.__version__
```

```python
import pandas as pd
athletes = pd.read_csv('athletes.csv')
athletes.head()
```

```python
athletes = athletes[athletes['sport'].isin(['wrestling', 'badminton', 'triathlon', 'gymnastics'])]
athletes.head()
```

# Relational Plots - Scatter and Line

```python
sns.relplot(data=athletes, x='weight', y='height')
```

```python
sns.relplot(data=athletes, x='weight', y='height', col='sport')
```

```python
sns.relplot(data=athletes, x='weight', y='height', col='sport', col_wrap=2, hue='sex')
```

```python
sns.relplot(data=athletes, x='weight', y='height', col='sport', col_wrap=2, hue='sex',
           kind='line')
```

```python
df = pd.read_csv('weather.csv') 
aug = df[df['MONTH'] == 8]
aug.head()
```

```python
# By default it plots the average temperature for every hour
# We also see the confidence interval for the average as the shaded ares

sns.relplot(data=aug, x='TIME', y='TEMP', kind='line')
```

# Visualizing Distributions of Data

```python
# Univariate 

sns.displot(data=athletes, x="sport");
```

```python
sns.displot(data=athletes, x="sport",hue='sex');
```

```python
# Bivariate

sns.displot(data=athletes, x="sport", y="weight", hue='sex');
```

# Categorical Plots

```python
sns.catplot(data=athletes, x="sport", y="height");
```

```python
sns.catplot(data=athletes, x="sport", y="weight", kind='swarm');
```

```python
sns.catplot(data=athletes, x="sport", y="weight", kind='box');
```

```python
sns.catplot(data=athletes, x="sport", y="weight", kind='boxen');
```

```python
sns.catplot(data=athletes, x="sport", y="weight", kind='violin', palette="pastel");
```

# Styles

```python
sns.displot(data=athletes, x="sport",hue='sex');
```

```python
sns.set_style('white')
sns.displot(data=athletes, x="sport",hue='sex');
```

```python
sns.set_style('whitegrid')
sns.displot(data=athletes, x="sport",hue='sex');
```

```python
sns.set_style('darkgrid')

sns.displot(data=athletes, x="sport",hue='sex');
```

```python
sns.set_style('ticks')
sns.set_context('poster', font_scale = 1)
sns.displot(data=athletes, x="sport",hue='sex',aspect=5/2);
```

```python
sns.set_theme()
sns.displot(data=athletes, x="sport",hue='sex');
```

```python
sns.set_palette("Blues")
sns.set_style('whitegrid')
sns.catplot(data=athletes, x="sport", y="weight", kind='violin');
```
